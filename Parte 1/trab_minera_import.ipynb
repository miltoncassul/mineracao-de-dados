# Eliminação de tuplas duplicadas

```python
df.drop_duplicates(subset=['DELEGACIA_NOME', 'ANO_BO', 'NUM_BO'], inplace=True)
df.reset_index(drop=True,inplace=True)
```

```python
df['HORARIOOCORRENCIA'] = pd.to_datetime(df.DATAOCORRENCIA + ' ' + df.HORAOCORRENCIA)
```

```python
df['DATAOCORRENCIA'] = pd.to_datetime(df['DATAOCORRENCIA'], format= '%Y-%m-%d') #%d-%m-%Y')
```

```python
#Após a remoção de tuplas duplicadas
depois_apagar = df.shape[0]
print("Quantidade de tuplas: ",df.shape[0])
```
